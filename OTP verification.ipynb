```python
from tkinter import *
```

```python
import random
import smtplib     # SMTP
import tkinter as tk  # GUI
from tkinter import messagebox    # to display message
```

```python
# Global variable to track whether OTP has been sent
otp_sent=False
```

```python
# generate a random 6-digit OTP
def generate_otp():
    otp=""
    for i in range(6):
        otp+=str(random.randint(0,9))
    return otp
# send OTP
def send_otp():
    global otp_sent   
    email_address ="vennela6575@gmail.com"
    email_password = "rzmd afuv djek mrop"
    recipient_email = email_entry.get()
    subject = "OTP"
 

    if not recipient_email or not subject:
        messagebox.showerror("Error", "Please enter valid email ID")       # error messsage if email-id is invalid
        return

    try:                                       # to test block of code for errors
        global OTP
        OTP = generate_otp()
        otp_msg = f"Subject: {subject}\n\n{OTP} is your OTP"

        s = smtplib.SMTP('smtp.gmail.com', 587)                                # creates SMTP session
        s.starttls()                                                           # start TLS for security
        s.login(email_address, email_password)                          # Authentication
        s.sendmail(email_address, recipient_email, otp_msg)            # sending the mail
        s.quit()                  # terminating the session
        messagebox.showinfo("Success", "OTP sent successfully to your email.")

        # Disable the "Send OTP" button
        send_button.config(state="disabled")
        otp_sent = True

    
        # Show the OTP entry and verify button
        otp_frame.pack()
    except smtplib.SMTPAuthenticationError:                            # to handle exception/error
        messagebox.showerror("Error", "SMTP Authentication Error.\n Please check your email and password.")
    except Exception as e:
        messagebox.showerror("Error",f"An error occured: {str(e)}")           # error raised other than authentication

# Verify OTP
def verify_otp():
    user_input = otp_entry.get()

    if user_input == OTP:
        messagebox.showinfo("Success", "OTP Verified")
        window.destroy()
    else:
        messagebox.showerror("Error", "Please check your OTP again")
 

# Resend OTP
def resend_otp():
    send_otp()

  # Re-enable the "send OTP" button after resending
    send_button.config(state="normal")                 # configure: to modify widgets options




# Create a GUI window
window = tk.Tk()    
window.title("OTP Verification")    # title of window created
window.geometry("400x250")          # shape of the window

# Email Entry
email_label = tk.Label(window,text="Enter your Email:")   
email_label.pack(pady=5)
email_entry = tk.Entry(window)
email_entry.pack(pady=5)

# send OTP button
send_button = tk.Button(window,text="Send OTP",activebackground="green", 
                   activeforeground="white" ,command=send_otp)
send_button.pack(pady=5)

# Frame to contain OTP entry and verify button(initially hidden)
otp_frame = tk.Frame(window)
# otp_frame.pack_forget()


# OTP Entry
otp_label = tk.Label(otp_frame, text="Enter your OTP:")
otp_label.pack(pady=5)
otp_entry = tk.Entry(otp_frame)
otp_entry.pack(pady=5)

# Verify OTP button
verify_button = tk.Button(otp_frame,text="Verify OTP",bg="light green", 
                   fg="black",command=verify_otp)
verify_button.pack(side=LEFT,padx=5,pady=5)

# Resend OTP button
resend_button= tk.Button(otp_frame,text="Resend OTP",bg="light blue", 
                   fg="black",command=resend_otp)
resend_button.pack(side=RIGHT,padx=5,pady=5)


window.mainloop()   # to run the application 
```
